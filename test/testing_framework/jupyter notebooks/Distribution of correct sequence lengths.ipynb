# Distribution of correct sequence lengths

This notebook is designed to get a distribution of all correct sequence lengths

```python
import matplotlib.pyplot as plt
import array
import os
import sys
from typing import Tuple
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from statistics import mean
from testing_utils import define_data, generate_truth_set
```

```python
datasets = define_data()
correct_sequences = generate_truth_set(datasets[0])
print(len(correct_sequences))
```

```python
int_list = []
correct_sequences_lengths = []
for j, i in enumerate(correct_sequences):
    correct_sequences_lengths.append(len(i))
    int_list.append(j)
plt.plot(int_list, correct_sequences_lengths, linestyle="",marker="o")
print(max(correct_sequences_lengths))
print(min(correct_sequences_lengths))
print(mean(correct_sequences_lengths))
```
