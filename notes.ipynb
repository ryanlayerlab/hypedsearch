# PPMs: what and why?

PPM is a essentially relative error:

$$
\text{ppm} = \frac{|\text{observed mass} - \text{theoretical mass} | }{|\text{theoretical mass} |} \cdot 10^6
$$

E.g., say you measure an ion with a mass of 500.0020 Da but the theoretical (expected) mass is 500 Da, then

$$
\text{ppm} = \frac{|500.0020 - 500 | }{|500 |} \cdot 10^6 = \frac{0.0020}{500} \cdot 10^6 = 4
$$

```python
def dalton_change_in_mass_for_given_ppm(mass_da, ppm):
    return ppm*mass_da*(10**(-6))

_, axs = fig_setup(1, 1)

ppm = 5
masses = np.arange(100, 2000, 100)
fcn = np.vectorize(dalton_change_in_mass_for_given_ppm)
_=axs[0].plot(
    masses, fcn(mass_da=masses, ppm=4)
)
finalize(axs)
set_title_axes_labels(
    axs[0], ylabel="Change in mass (Da)", 
    xlabel="Mass (Da)", 
    title=f"Change in mass corresponding to {ppm} ppm change"
)
```
